# GC Example 24: New Clusters Beyond Harris (2010)

**EPS Research — Milky Way GC Corpus v1.3.2**

17 clusters discovered after Harris (2010) are in the corpus
from Vasiliev & Baumgardt (2021). They have Gaia data but no Harris photometry.

**Corpus:** Flynn (2026), Zenodo DOI: 10.5281/zenodo.19907766  
**Sources:** Harris (1996/2010), Vasiliev & Baumgardt (2021), Baumgardt et al. (2023), Schiavon et al. (2024) APOGEE DR17  
**Dependencies:** Python 3, numpy, matplotlib

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'harris_gc_corpus_v1.3.2.jsonl': 'https://zenodo.org/records/21093446/files/harris_gc_corpus_v1.3.2.jsonl',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

clusters = []
with open('harris_gc_corpus_v1.3.2.jsonl') as f:
    for line in f:
        clusters.append(json.loads(line))
print(f"Total clusters: {len(clusters)}")
```

```python
new_clusters=[c for c in clusters
              if not (c.get('metallicity') and c['metallicity'].get('feh') is not None)
              and c.get('gaia_edr3')]
print(f"Post-Harris clusters (Gaia only): {len(new_clusters)}")
print(f"\n{'Cluster':<20} {'mu_alpha':>10} {'mu_delta':>10} {'N_Gaia':>8}")
print('-'*52)
for c in new_clusters:
    g=c['gaia_edr3']
    print(f"{c['cluster_id']:<20} {g.get('mu_alpha_mas_yr') or 0:>10.3f} "
          f"{g.get('mu_delta_mas_yr') or 0:>10.3f} {int(g.get('n_members_gaia') or 0):>8}")
print(f"\nNote: These clusters have no Harris photometric parameters.")
print(f"They contribute Gaia EDR3 proper motions only.")
print(f"5 of these also have Baumgardt N-body data.")
```
