# GC Example 16: Selecting GCs for N-body Follow-up

**EPS Research — Milky Way GC Corpus v1.3.1**

Which clusters are best suited for N-body simulation follow-up?
Criteria: well-characterized mass, moderate relaxation time, sufficient members.

**Corpus:** Flynn (2026), Zenodo DOI: 10.5281/zenodo.19907765  
**Sources:** Harris (1996/2010), Vasiliev & Baumgardt (2021), Baumgardt et al. (2023), Schiavon et al. (2024) APOGEE DR17  
**Dependencies:** Python 3, numpy, matplotlib

```python
# ── Colab setup: download corpus from Zenodo ──────────────────
import os, urllib.request
CORPORA = {
    'harris_gc_corpus_v1.3.1.jsonl': 'https://zenodo.org/records/19907765/files/harris_gc_corpus_v1.3.1.jsonl',
}
for filename, url in CORPORA.items():
    if not os.path.exists(filename):
        print(f"Downloading {filename}...")
        urllib.request.urlretrieve(url, filename)
        print(f"  ✓ {filename}")
    else:
        print(f"  Already present: {filename}")
print("Ready.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

clusters = []
with open('harris_gc_corpus_v1.3.1.jsonl') as f:
    for line in f:
        clusters.append(json.loads(line))
print(f"Total clusters: {len(clusters)}")
```

```python
candidates=[]
for c in clusters:
    b=c.get('baumgardt2023',{})
    if not b or not b.get('mass_msun'): continue
    mass=b['mass_msun']; log_trh=b.get('log_trh_yr',0)
    n_rv=b.get('n_rv',0); n_pm=b.get('n_pm',0)
    score=0
    if 1e4<mass<1e6: score+=2
    if log_trh>9.5: score+=2
    if n_rv>100: score+=1
    if n_pm>1000: score+=1
    candidates.append({'cluster':c['cluster_id'],'mass':mass,
                       'log_trh':log_trh,'n_rv':n_rv,'n_pm':n_pm,'score':score})

top=sorted(candidates,key=lambda x:-x['score'])[:10]
print("Top 10 GCs for N-body follow-up:")
print(f"{'Cluster':<15} {'Mass':>10} {'log t_rh':>9} {'N_rv':>6} {'N_pm':>7} {'Score':>6}")
print('-'*58)
for c in top:
    print(f"{c['cluster']:<15} {c['mass']:>10,.0f} {c['log_trh']:>9.2f} "
          f"{c['n_rv']:>6} {c['n_pm']:>7} {c['score']:>6}")
```
