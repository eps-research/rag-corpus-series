# IntZ Example 2: Redshift Distribution
**EPS Research IntZ Kinematic Corpus v1.0**

Compare redshift distributions of KROSS and KMOS3D sub-samples.

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
z_kross  = [g['redshift']['z_spec'] for g in galaxies
             if g['metadata']['survey']=='KROSS']
z_kmos3d = [g['redshift']['z_spec'] for g in galaxies
             if g['metadata']['survey']=='KMOS3D']

fig, ax = plt.subplots(figsize=(8,4))
ax.hist(z_kross,  bins=30, alpha=0.7, color='steelblue',
        label=f'KROSS (n={len(z_kross)})')
ax.hist(z_kmos3d, bins=30, alpha=0.7, color='darkorange',
        label=f'KMOS3D (n={len(z_kmos3d)})')
ax.axvline(np.median(z_kross),  color='steelblue',  ls='--', lw=1.5)
ax.axvline(np.median(z_kmos3d), color='darkorange', ls='--', lw=1.5)
ax.set_xlabel('Spectroscopic Redshift')
ax.set_ylabel('N galaxies')
ax.set_title('IntZ Corpus: Redshift Distribution by Survey')
ax.legend()
plt.tight_layout()
plt.savefig('intz_nb2_redshift_dist.png', dpi=120)
plt.show()
print(f"KROSS median z = {np.median(z_kross):.3f}")
print(f"KMOS3D median z = {np.median(z_kmos3d):.3f}")
```
