Correlate omega with stellar mass for Tier-1 galaxies.

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'intz_corpus_v1b.json': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b.json',
        'intz_corpus_v1b_flat.csv': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b_flat.csv',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
data = []
for g in galaxies:
    if not g['omega']['omega_available']: continue
    om = g['omega']['omega_value_rad_gyr']
    sp = g.get('stellar_properties', {})
    lm = sp.get('log_mass_msun')
    if om is not None and lm:
        data.append((lm, om))

lm_v, om_v = zip(*data)

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(lm_v, om_v, c='firebrick', alpha=0.4, s=15)
ax.axhline(0, color='gray', ls='-', lw=1, alpha=0.5)
ax.set_xlabel('log M* (M☉)')
ax.set_ylabel('ω (rad/Gyr)')
ax.set_title('IntZ Tier-1: Omega vs Stellar Mass')
plt.tight_layout()
plt.savefig('intz_nb12_omega_mass.png', dpi=120)
plt.show()

from scipy.stats import spearmanr
rho, p = spearmanr(lm_v, om_v)
print(f"Spearman rho = {rho:.3f}, p = {p:.3f}")
```
