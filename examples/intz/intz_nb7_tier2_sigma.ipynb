# IntZ Example 7: Tier-2 Integrated Dispersions
**EPS Research IntZ Kinematic Corpus v1.0**

Explore integrated velocity dispersions for KMOS3D Tier-2 galaxies. Note: these are integrated line widths, NOT beam-corrected σ₀.

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
t2_kmos = [g for g in galaxies
            if g['metadata']['survey']=='KMOS3D']

sigma = [g['kinematics'].get('sigma0_obs_kms') or
         g['kinematics'].get('sigma_tot_kms')
         for g in t2_kmos]
sigma = [s for s in sigma if s and 0 < s < 400]

fig, ax = plt.subplots(figsize=(7,4))
ax.hist(sigma, bins=30, color='darkorange', alpha=0.8, edgecolor='k', lw=0.3)
ax.axvline(np.median(sigma), color='black', ls='--', lw=2,
           label=f'Median = {np.median(sigma):.0f} km/s')
ax.set_xlabel('σ_int (km/s)  [integrated line width, NOT σ₀]')
ax.set_ylabel('N galaxies')
ax.set_title(f'KMOS3D Integrated Velocity Dispersions (N={len(sigma)})')
ax.legend()
plt.tight_layout()
plt.savefig('intz_nb7_sigma_t2.png', dpi=120)
plt.show()
print("WARNING: These are integrated line widths, not beam-corrected intrinsic dispersions.")
print("Do not mix with Tier-1 sigma0 values.")
```
