# IntZ Example 5: Omega Distribution
**EPS Research IntZ Kinematic Corpus v1.0**

Distribution of omega values for 166 Tier-1 KROSS galaxies under the Flynn & Cannaliato (2025) correction.

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'intz_corpus_v1b.json': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b.json',
        'intz_corpus_v1b_flat.csv': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b_flat.csv',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
import statistics
omega = [g['omega']['omega_value_rad_gyr'] for g in galaxies
         if g['omega']['omega_available']
         and g['omega']['omega_value_rad_gyr'] is not None]

fig, ax = plt.subplots(figsize=(8,4))
ax.hist(omega, bins=25, color='firebrick', alpha=0.8, edgecolor='darkred')
ax.axvline(statistics.median(omega), color='black', ls='--', lw=2,
           label=f'Median = {statistics.median(omega):.2f} rad/Gyr')
ax.axvline(0, color='gray', ls='-', lw=1, alpha=0.5)
ax.axvline(7.06, color='green', ls=':', lw=2, label='z=0 SPARC +7.06')
ax.set_xlabel('ω (rad/Gyr)')
ax.set_ylabel('N galaxies')
ax.set_title(f'IntZ Tier-1 Omega Distribution (N={len(omega)}, 100% negative)')
ax.legend()
plt.tight_layout()
plt.savefig('intz_nb5_omega_dist.png', dpi=120)
plt.show()
print(f"Median: {statistics.median(omega):.3f} rad/Gyr")
print(f"All negative: {sum(1 for o in omega if o<0)}/{len(omega)}")
```
