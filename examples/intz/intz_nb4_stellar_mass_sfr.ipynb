# IntZ Example 4: Star-Forming Main Sequence
**EPS Research IntZ Kinematic Corpus v1.0**

Plot stellar mass vs SFR for both surveys.

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'intz_corpus_v1b.json': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b.json',
        'intz_corpus_v1b_flat.csv': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b_flat.csv',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
fig, ax = plt.subplots(figsize=(7,5))
for survey, color in [('KROSS','steelblue'),('KMOS3D','darkorange')]:
    lm, sfr = [], []
    for g in galaxies:
        if g['metadata']['survey'] != survey: continue
        sp = g.get('stellar_properties', {})
        m = sp.get('log_mass_msun') or sp.get('log_mstar_msun')
        s = sp.get('sfr_msun_yr')
        if m and s and s > 0:
            lm.append(m); sfr.append(np.log10(s))
    ax.scatter(lm, sfr, c=color, alpha=0.3, s=8,
               label=f'{survey} (n={len(lm)})')

ax.set_xlabel('log M* (M☉)')
ax.set_ylabel('log SFR (M☉/yr)')
ax.set_title('IntZ: Star-Forming Main Sequence')
ax.legend()
plt.tight_layout()
plt.savefig('intz_nb4_main_seq.png', dpi=120)
plt.show()
```
