# IntZ Example 10: RAG JSONL Demo
**EPS Research IntZ Kinematic Corpus v1.0**

Demonstrate LLM RAG ingestion using the JSONL format. Each line is a self-describing galaxy record.

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'intz_corpus_v1b.json': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b.json',
        'intz_corpus_v1b_flat.csv': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b_flat.csv',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import os
import json

JSONL_PATH = os.path.expanduser('~/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.jsonl')

# Read first 5 records
records = []
with open(JSONL_PATH) as f:
    for i, line in enumerate(f):
        if i >= 5: break
        records.append(json.loads(line))

print(f"Sample RAG summaries (first 5 records):")
print("-" * 60)
for r in records:
    print(r.get('rag_summary', 'No summary'))
    print()
```

```python
# Demonstrate keyword search over JSONL
print("Searching for galaxies with omega < -20 rad/Gyr...")
hits = []
with open(JSONL_PATH) as f:
    for line in f:
        g = json.loads(line)
        om = g['omega'].get('omega_value_rad_gyr')
        if om and om < -20:
            hits.append(g['rag_summary'][:100])

print(f"Found {len(hits)} matches:")
for h in hits[:5]:
    print(f"  {h}")
```
