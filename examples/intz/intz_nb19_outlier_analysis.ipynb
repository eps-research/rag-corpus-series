# IntZ Example 19: Omega Outlier Analysis
**EPS Research IntZ Kinematic Corpus v1.0**

Deep-dive into the 4 omega outlier galaxies (|ω| > 3σ threshold).

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
import statistics

omega_all = [g['omega']['omega_value_rad_gyr'] for g in galaxies
             if g['omega']['omega_available']
             and g['omega']['omega_value_rad_gyr'] is not None]
med = statistics.median(omega_all)
std = statistics.stdev(omega_all)
threshold = abs(med) + 3*std
print(f"Median: {med:.3f}, Stdev: {std:.3f}, 3σ threshold: {threshold:.3f}")

outliers = [(g['identifiers'],
             g['omega']['omega_value_rad_gyr'],
             g['redshift']['z_spec'],
             g['kinematics'].get('Vc_kms'),
             g['stellar_properties'].get('log_mass_msun'))
            for g in galaxies
            if g['omega']['omega_available']
            and g['omega'].get('omega_outlier_flag')]

print(f"\nOutlier galaxies (N={len(outliers)}):")
print(f"{'ID':20s} {'omega':>10s} {'z':>6s} {'Vc':>8s} {'logM*':>6s}")
print("-"*55)
for ids, om, z, vc, lm in sorted(outliers, key=lambda x: x[1]):
    gid = ids.get('kid', ids.get('id', '?'))
    print(f"{str(gid):20s} {om:10.3f} {z:6.3f} {str(vc)[:6]:>8s} {str(lm)[:5]:>6s}")
```
