# IntZ Example 6: Three Canonical Examples
**EPS Research IntZ Kinematic Corpus v1.0**

Detailed inspection of KID-141, GS4_33971, and KID-522 — the three canonical example galaxies for the IntZ corpus.

```python
# ── Colab setup: auto-download corpus from Zenodo ─────────────
import os, sys
IN_COLAB = 'google.colab' in sys.modules
if IN_COLAB:
    import urllib.request
    CORPORA = {
        'intz_corpus_v1b.json': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b.json',
        'intz_corpus_v1b_flat.csv': 'https://zenodo.org/records/20453189/files/intz_corpus_v1b_flat.csv',
    }
    for filename, url in CORPORA.items():
        if not os.path.exists(filename):
            print(f"Downloading {filename}...")
            urllib.request.urlretrieve(url, filename)
            print(f"  ✓ {filename}")
        else:
            print(f"  Already present: {filename}")
    print("Ready.")
else:
    print("Running locally — corpus files loaded from working directory.")
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
targets = {
    'KID-141':    lambda g: g['identifiers'].get('kid')==141,
    'GS4_33971':  lambda g: g['identifiers'].get('id')=='GS4_33971',
    'KID-522':    lambda g: g['identifiers'].get('kid')==522,
}

for label, fn in targets.items():
    g = next((g for g in galaxies if fn(g)), None)
    if not g:
        print(f"{label}: NOT FOUND")
        continue
    sp = g.get('stellar_properties', {})
    kin = g['kinematics']
    om = g['omega']
    red = g['redshift']
    print(f"\n{'='*50}")
    print(f"{label}")
    print(f"  z_spec:  {red['z_spec']:.4f}")
    print(f"  log M*:  {sp.get('log_mass_msun','?')}")
    print(f"  SFR:     {sp.get('sfr_msun_yr','?')} M☉/yr")
    print(f"  Vc:      {kin.get('Vc_kms','?')} km/s")
    print(f"  sigma0:  {kin.get('sigma0_kms','?')} km/s")
    print(f"  v/sigma: {kin.get('v_over_sigma','?')}")
    print(f"  omega:   {om.get('omega_value_rad_gyr','N/A')} rad/Gyr")
    print(f"  RAG:     {g.get('rag_summary','')[:100]}")
```
