# IntZ Example 1: Load and Explore the IntZ Corpus
**EPS Research IntZ Kinematic Corpus v1.0 — Flynn D.C. (2026)**

This notebook demonstrates how to load the IntZ corpus and inspect its structure, survey breakdown, and tier statistics.

```python
import json
import numpy as np
import matplotlib.pyplot as plt

# Load IntZ corpus
with open('/home/david/Documents/RAG Project/Z=2 RAG/Zenodo/intz_corpus_v1b.json') as f:
    data = json.load(f)
galaxies = data['galaxies']
print(f"Loaded {len(galaxies)} galaxies")
```

```python
# Survey and tier breakdown
from collections import Counter
surveys = Counter(g['metadata']['survey'] for g in galaxies)
tiers   = Counter(g['metadata']['quality_tier'] for g in galaxies)
print("By survey:", dict(surveys))
print("By tier:  ", dict(tiers))

# Redshift range
z_all = [g['redshift']['z_spec'] for g in galaxies]
print(f"z range: {min(z_all):.3f} - {max(z_all):.3f}, median: {np.median(z_all):.3f}")
```

```python
# Inspect one galaxy record
g0 = galaxies[0]
print("Top-level keys:", list(g0.keys()))
print("\nSample RAG summary:")
print(g0['rag_summary'])
```

```python
# Count omega-ready galaxies
n_omega = sum(1 for g in galaxies if g['omega']['omega_available'])
print(f"Omega-ready (Tier-1): {n_omega}/{len(galaxies)}")
```
