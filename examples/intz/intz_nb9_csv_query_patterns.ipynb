# IntZ Example 9: CSV Query Patterns
**EPS Research IntZ Kinematic Corpus v1.0**

Demonstrate common pandas query patterns on the flat CSV.

```python
import pandas as pd
import numpy as np

CSV_PATH = '~/Documents/RAG Project/Z=2 RAG/intz_corpus_v1b_flat.csv'
df = pd.read_csv(CSV_PATH)
print(f"Loaded {len(df)} rows, {len(df.columns)} columns")
print(df.columns.tolist())
```

```python
# Filter Tier-1 with omega
t1 = df[df['quality_tier']==1]
print(f"Tier-1: {len(t1)} galaxies")
print(t1[['galaxy_id','z_spec','Vc_kms','omega_value_rad_gyr']].head(10))
```

```python
# High v/sigma galaxies
high_vos = df[df['v_over_sigma'] > 5].copy()
print(f"v/sigma > 5: {len(high_vos)} galaxies")
print(high_vos[['galaxy_id','survey','z_spec','v_over_sigma']].head(10))
```

```python
# KMOS3D massive galaxies at z>2
massive_highz = df[(df['survey']=='KMOS3D') & (df['z_spec']>2.0)]
print(f"KMOS3D z>2: {len(massive_highz)} galaxies")
```
