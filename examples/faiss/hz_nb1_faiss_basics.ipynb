# FAISS Semantic Search — Basics
**EPS Research RAG Corpus Series**

This notebook demonstrates natural-language similarity search across all five EPS astrophysical corpora using the live MCP server REST endpoint. No installation required beyond `requests`.

**Live endpoint:** https://dflynn5656-astro-rag-mcp.hf.space

**DOI:** 10.5281/zenodo.21154451

```python
import requests
import pandas as pd

BASE = 'https://dflynn5656-astro-rag-mcp.hf.space/api'

def semantic_search(corpus, query, top_k=5):
    r = requests.get(f'{BASE}/semantic_search',
                     params={'corpus': corpus, 'query': query, 'top_k': top_k})
    data = r.json()['data']
    return pd.DataFrame(data['results'])

print('Ready.')
```

## 1. Query the v7 HI Rotation Curve Corpus

```python
# Find galaxies similar to DDO154 — a classic low-mass dwarf irregular
results = semantic_search('v7', 'dwarf irregular low surface brightness low mass', top_k=10)
print(results.to_string(index=False))
```

## 2. Query the Globular Cluster Corpus

```python
# Find metal-poor outer halo clusters
results = semantic_search('gc', 'metal poor outer halo low metallicity', top_k=10)
print(results.to_string(index=False))
```

## 3. Query the IntZ Corpus (z~0.4-2.7)

```python
# Find high-redshift rotating disks from KROSS
results = semantic_search('intz', 'high redshift rotating disk KROSS massive star forming', top_k=10)
print(results.to_string(index=False))
```

## 4. Query the Z1 ALPINE Corpus (z~4-6)

```python
# Find confirmed rotators in the ALPINE survey
results = semantic_search('z1', 'confirmed rotator ALPINE CII high redshift', top_k=10)
print(results.to_string(index=False))
```

## 5. Understanding Scores
The `score` is the L2 distance in 384-dimensional embedding space.
**Lower score = more similar.** Scores below ~1.3 are typically good semantic matches.
Scores above ~1.5 indicate the query is finding less semantically related objects.

```python
# Compare scores for a tight vs broad query
tight  = semantic_search('v7', 'NGC3198 spiral rotation curve', top_k=5)
broad  = semantic_search('v7', 'galaxy', top_k=5)
print('Tight query scores:', tight['score'].round(3).tolist())
print('Broad query scores:', broad['score'].round(3).tolist())
```
