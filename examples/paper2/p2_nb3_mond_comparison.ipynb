# Baryonic Validation 3: Omega vs MOND ⭐ Published Results
**Flynn (2026), 84-Galaxy Baryonic Validation (Flynn 2026)**

On the same baryonic target (Vbary), the omega correction outperforms MOND
in 82/84 galaxies (25.45 vs 60.57 km/s; 2.4× improvement).

**Key point:** MOND is designed to predict Vobs from baryonic mass.
Omega is designed to recover Vbary from observed kinematics.
They are complementary, not competing.
